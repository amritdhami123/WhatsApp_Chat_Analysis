<h3>1) Preprocessing</h3>

```python
# For 12-hour format (most common)
pattern = r"\d{1,2}/\d{1,2}/\d{4},\s\d{1,2}:\d{2}\u202f[ap]m\s-\s"
```

```python
messages = re.split(pattern, data)[1:]
len(message)
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({"user_message": messages, "message_date": dates})

df["message_date"] = df["message_date"].str.replace(r"\s[–-]\s$", "", regex=True)

df["datetime"] = pd.to_datetime(df["message_date"], format="%d/%m/%Y, %I:%M\u202f%p")

df = df.drop(columns=["message_date"]).rename(columns={"datetime": "date"})


print(df.head())
```

```python
df.shape # Numbers of message is in the group. 
```

```python
users = []
messages = []
for message in df['user_message']:  # Fixed: removed space in column name
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')  # Fixed: added dot before append
        messages.append(entry[0])

df['user'] = users
df['message'] = messages  
df.drop(columns=['user_message'], inplace=True)  
df.head()  
```

```python
df['year'] = df['date'].dt.year #Exctracting the year from dataset.
```
